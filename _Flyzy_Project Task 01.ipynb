```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot for the 'Flight_Distance' variable
plt.figure(figsize=(8, 6))
sns.boxplot(x=Data['Flight_Distance'])
plt.title('Boxplot of Flight Distance')
plt.xlabel('Flight Distance (Scaled)')
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) of Flight_Distance
Q1 = Data['Flight_Distance'].quantile(0.25)
Q3 = Data['Flight_Distance'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
filtered_Data = Data[(Data['Flight_Distance'] >= lower_bound) & (Data['Flight_Distance'] <= upper_bound)]

# Display the boxplot of Flight_Distance after removing outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=filtered_Data['Flight_Distance'])
plt.title('Boxplot of Flight Distance (Outliers Removed)')
plt.xlabel('Flight Distance (Scaled)')
plt.show()
```
